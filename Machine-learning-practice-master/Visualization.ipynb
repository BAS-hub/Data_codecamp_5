```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('kc_house_data.csv')
df
```

```python
fig = plt.figure(figsize=(10,8))
sns.distplot(df['price'])
```

```python
fig = plt.figure(figsize=(10,8))
sns.boxplot(df['price'],orient = 'v')
```

```python
fig = plt.figure(figsize=(10,8))
sns.countplot(df['bedrooms'])
```

```python
fig = plt.figure(figsize=(10,8))
sns.countplot(df['grade'])
```

```python
fig = plt.figure(figsize=(30,20))
plt.title('house price by squre living')
plt.xlabel('squre living')
plt.ylabel('house price')
sns.barplot(x='sqft_living',y='price',data=df)
```

```python
fig = plt.figure(figsize=(30,20))
plt.title('house price by grade')
plt.xlabel('grade')
plt.ylabel('house price')
sns.barplot(x='grade',y='price',data=df)
```

```python
fig = plt.figure(figsize=(30,20))
plt.title('house price by floors')
plt.xlabel('floors')
plt.ylabel('house price')
sns.barplot(x='floors',y='price',data=df)
```

```python
df.groupby('floors').mean()['price']
```

```python
df_corr = df[['price','floors','sqft_living','bedrooms','bathrooms','grade']]
sns.pairplot(df_corr)
```

```python
sns.heatmap(df_corr.corr())
```
