```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
data
```

```python
data.feature_names
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data.data ,columns = data.feature_names)
df
```

```python
label = data.target ## Classification
label
```

```python
len(label)
```

```python
df['label'] =  label
```

```python
df.sample(10)
```

```python
df['label_name'] = ['malinant' if x==0 else 'benign' for x in df['label']]
```

```python
df
```

```python
df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df.drop(['label','label_name'],axis=1),df['label'],test_size=0.5,random_state=1)
```

```python
print('x_train shape = ',x_train.shape)
print('y_train shape = ',y_train.shape)
print('x_test shape = ',x_test.shape)
print('y_test shape = ',y_test.shape)
```

```python
## Default parameter
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
nb.fit(x_train,y_train)
```

```python
predicted = nb.predict(x_test)
predicted
```

```python
from sklearn.metrics import confusion_matrix ,accuracy_score ,f1_score ,precision_score ,recall_score,classification_report
```

```python
confusion_matrix(y_test,predicted)
```

```python
print(classification_report(y_test,predicted))
```
