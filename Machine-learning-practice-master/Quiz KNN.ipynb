```python
df = pd.read_csv('Wine.csv',header = None)
df.sample(5)
```

```python
df.columns = ['wine class','Malic acid','Alcohol','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
df.head(5)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(36, 24))
sns.boxplot(x='wine class' ,y='OD280/OD315 of diluted wines', data=df,ax=axes[0,0])
sns.boxplot(x='wine class' ,y='Flavanoids', data=df,ax=axes[0,1])
sns.boxplot(x='wine class' ,y='Total phenols', data=df,ax=axes[1,0])
sns.boxplot(x='wine class' ,y='Hue', data=df,ax=axes[1,1])
plt.rcParams.update({'font.size': 50})
plt.tight_layout()
```

```python
##Min-Max normalization -  [0,1]
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
```

```python
df_minmax = min_max_scaler.fit_transform(df.drop(['wine class'],axis=1))
df_minmax
```

```python
df_minmax.shape
```

```python
new_df = pd.DataFrame(df_minmax,columns =df.columns[1:])
new_df
```

```python
## Split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(new_df,df['wine class'], test_size=0.3, random_state=100)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
## KNN Model training 
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(x_train,y_train)
```

```python
predicted = knn.predict(x_test)
```

```python
## Model Evaluation
from sklearn.metrics import confusion_matrix , accuracy_score ,f1_score ,recall_score ,precision_score,classification_report
```

```python
confusion_matrix(y_test,predicted)
```

```python
print(classification_report(y_test, predicted))
```

```python
## Parametor tuning random k between 1-30
accuracy_list = []
for i in range(1,30) :
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(x_train,y_train)
    predicted_i = knn.predict(x_test)
    accuracy_list.append(accuracy_score(y_test,predicted_i))
```

```python
plt.figure(figsize = (12,8))
plt.plot(range(1,30),accuracy_list ,color='black',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=7)
plt.xlabel('Number of K')
plt.ylabel('Accuracy')
plt.rcParams.update({'font.size': 10})
```

```python
from sklearn.model_selection import GridSearchCV ## Parameter tuning
parameters = {'n_neighbors': np.arange(1,10,1)}
knn_best = GridSearchCV(knn, parameters, cv=2,scoring = 'roc_auc_ovr')
knn_best.fit(x_train,y_train)
knn_best.best_estimator_
```

```python
predicted_best = knn_best.predict(x_test)
predicted_best
```

```python
print(classification_report(y_test, predicted_best))
```
