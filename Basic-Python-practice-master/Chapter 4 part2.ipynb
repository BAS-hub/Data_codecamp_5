```python
## การจัดการข้อมูลประเภท Missing Values
import numpy as np
import pandas as pd
df = pd.read_csv('D:\Automobile_data.CSV', na_values ={'normalized-losses':['?'], 'stroke':['?'],'horsepower':['?'],'peak-rpm':['?'],'price':['?']}) ## เเปลงข้อมูลที่หายไปเป็น NaN
df ## เครื่องหมาย ? เเทนค่าที่ถูกบันทึกไว้ (ค่าว่างกลายเป็นเครื่องหมาย) ต้องเเทน ? ด้วยค่า NaN

```

```python
## df.dropna(axis=1),df.dropna(axis=0) ลบข้อมูล NaN ที่ปรากฏตามหลัก (Axis=1) หรือ ตามเเถว (Axis =0)
df.dropna(axis=0) ## save การเเก้ไขด้วย inplace = True
```

```python
df.fillna(value = 0) ##เเทนค่า NaN ด้วย 0 เเต่ใช้ได้กับกลุ่มข้อมูลที่เป็นตัวเลขเท่านั้น
```

```python
df['normalized-losses'].fillna(value = 0)  ##เเทนค่า NaN ด้วย 0 เฉพาะ colum
```

```python
avg = df['normalized-losses'].mean() ## เป็นวิธียอดฮิต เพื่อคงค่าเฉลี่ยของข้อมูลให้คงเดิม เเทนค่า NaN ด้วยค่าเฉลี่ย
print(avg)
df['normalized-losses'].fillna(value = avg) ## อย่าลืม inplace = True น่ะครับบบบ
```
