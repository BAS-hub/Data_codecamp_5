```python
## Real data for Seaborn (Caregorical and Numerical data !!!!!!!!!)
df2 = pd.read_csv('D:\Automobile_data.CSV', na_values ={'normalized-losses':['?'], 'stroke':['?'],'horsepower':['?'],'peak-rpm':['?'],'price':['?'],'num-of-doors':['?']}) ## เเปลงข้อมูลที่หายไปเป็น NaN
df2=df2.dropna(axis=0)
df2 ## เครื่องหมาย ? เเทนค่าที่ถูกบันทึกไว้ (ค่าว่างกลายเป็นเครื่องหมาย) ต้องเเทน ? ด้วยค่า NaN
```

```python
fg = plt.figure(figsize=(8,8)) ##เเก้ปัญหาเเกน x ทับกัน
sns.countplot(x='make',data=df2) ## countplot ไม่ต้องใส่เเกน y สำหรับ categorical plot
##หรือ  sns.countplot(x=df2['make'])
fg.autofmt_xdate() ## เฉียงเเกน x ให้ไม่ทับกัน
plt.show()
```

```python
fg = plt.figure(figsize=(8,8)) ##เเก้ปัญหาเเกน x ทับกัน
sns.countplot(y='body-style',data=df2) ## countplot ไม่ต้องใส่เเกน y สำหรับ categorical plot
##หรือ  sns.countplot(x=df2['make'])
##fg.autofmt_xdate() ## เฉียงเเกน x ให้ไม่ทับกัน
plt.show()
```

```python
### Seaborn Boxplot *** บอกความหมายทางสถิติได้
fg = plt.figure(figsize=(15,10))
df2['price']=pd.to_numeric(df2['price']) ###เปลี่ยนเป็น numerical
sns.boxplot(x ='make',y='price',data=df2) ## countplot ไม่ต้องใส่เเกน y สำหรับ categorical plot
fg.autofmt_xdate()
plt.show()
```

```python
fg = plt.figure(figsize=(10,8))
df2['price']=pd.to_numeric(df2['price']) ###เปลี่ยนเป็น numerical
sns.boxplot(x ='body-style',y='price',data=df2) ## countplot ไม่ต้องใส่เเกน y สำหรับ categorical plot
plt.show()
```

```python
## Seaborn barplot
fg = plt.figure(figsize=(5,5))
df2['price']=pd.to_numeric(df2['price']) ###เปลี่ยนเป็น numerical
sns.barplot(x ='body-style',y='price',data=df2) ## หาค่าเฉลี่ยเเต่ละกลุ่ม คล้ายๆ groupby ใน pandas
fg.autofmt_xdate()
plt.show()
```

```python
### Seaborn stripplot พล็อตจุดข้อมูลเป็นกลุ่มๆ
fg = plt.figure(figsize=(12,8))
df2['price']=pd.to_numeric(df2['price']) ###เปลี่ยนเป็น numerical
sns.stripplot(x ='body-style',y='horsepower',data=df2,hue='num-of-doors',dodge=True) ## hue เเบ่งประเภทภายในกราฟได้ dodge ใช้เเยกข้อมูลที่ hue
##fg.autofmt_xdate()
plt.show()
```

```python
### Seaborn swarmplot พล็อตจุดข้อมูลเป็นกลุ่มๆเหมาะสำหรับข้อมูลที่ทับกันบางส่วน
fg = plt.figure(figsize=(12,8))
df2['price']=pd.to_numeric(df2['price']) ###เปลี่ยนเป็น numerical
sns.swarmplot(x ='body-style',y='horsepower',data=df2,hue='num-of-doors',dodge=True) ## hue เเบ่งประเภทภายในกราฟได้ dodge ใช้เเยกข้อมูลที่ hue
##fg.autofmt_xdate()
plt.show()
```

```python
##Seaborn heatmap ต้องเปลี่ยน dataframe เป็น matrix form (ต้องไม่มี index ชี้ตำเเหน่ง row)
df2.head()
```

```python
new_df2 = pd.crosstab(df2['body-style'],df2['drive-wheels']) ## ตำเเหน่งเเรกคือ row ตำเเหน่งที่ 2 คือ colum,value = จำนวนที่นับได้
new_df2
```

```python
ax = sns.heatmap(new_df2,cmap='coolwarm',annot = True,square=True) ##cmap คือเปลี่ยนสีกราฟ annot ใส่ตัวเลขลงในตาราง
ax.set_aspect('equal')
```

```python
new_df3 = df2.pivot_table(index = 'body-style',columns = 'drive-wheels',values ='horsepower')
new_df3
```

```python
sns.heatmap(new_df3,cmap='coolwarm',annot = True,linecolor = 'white',linewidth=1.5)
```

```python
df3 = df2[['body-style','price']]
df3 = df3.dropna()
check = df3.groupby(['body-style']) #### นำ body-style มาเป็นหลักเเทน index
check2 = check.describe()['price'][['mean','max','std']] ## เเสดงข้อมูลเชิงสถิติของ price ในเเต่ละ body-style
check2
```

```python
sns.heatmap(check2,cmap='coolwarm',annot = True,linecolor = 'white',linewidth=1.5)
```

```python
#### Colleration (negative,positive)
df_corr = df2.corr() ## ตัดข้อมูลที่เป็น str ออกทั้งหมด
df_corr ## เทียบข้อมูลที่เป็นตัวเลขทุกๆ column
```

```python
fg = plt.figure(figsize=(12,8))
sns.heatmap(df_corr,cmap='coolwarm',annot = True,linecolor = 'white',linewidth=1.5) ## เส้นทะเเยงสีเเดงคือข้อมูลเทียบกันเอง
```

```python
sns.clustermap(df_corr,cmap='coolwarm',annot=True) ## นำข้อมูลที่ใกล้เคียงกันอยู่ด้วยกัน
```

```python
##Seaborn scatter plot (หลังจากรู้ความสัมพันธ์ระหว่างข้อมูล correlation เเล้วก็เลือกนำมา plot)
fig = plt.figure(figsize=(12,8))
sns.scatterplot(x='price',y='horsepower',data=df2,hue='num-of-doors')
```

```python
## Distributuion plot (คล้ายๆ histrogram)
sns.distplot(df2['price'],bins=10) ## มีเส้น KDE เเสดงความหนาเเน่นของข้อมูล ,bins = จำนวนเเท่งกราฟ
```
