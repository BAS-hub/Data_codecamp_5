```python
##Introduction mathplotlib
import numpy as np
import matplotlib.pyplot as plt ##plot graph พื้นฐาน
##มองเห็นได้ใน jupiter notebook
%matplotlib inline

x = [1,2,3,4,5,6,7,8,9]
x = np.array(x)
y = x**2
plt.plot(x,y,marker = 'o', linewidth = -1) ## marker = จุด , linewidth (+,-) ความหนาของเส้น
plt.xlabel('Number  1 - 9')
plt.ylabel('X axis square')
plt.title('Exponential Graph')
plt.show() ## เเสดงผลกราฟ (กรณีไม่ใช้ %matplotlib inline)
```

```python
x = np.arange(0,90,1)
y = np.sin(x) ## หน่วยเป็น เรเดียน(rad)
plt.plot(x,y,marker = '^')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sindusindual Graph')
plt.show()
```

```python
x = np.arange(0,90,1)
y = np.sin(x) ## หน่วยเป็น เรเดียน(rad)
plt.plot(x,y,ls='--',marker = 'o') ## ls (line style) เป็นเส้นประ  maker จุด
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sindusindual Graph')
plt.show()
```

```python
x = np.arange(0,20,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label = 'sin(x)',c='g') ## label เเสดงชื่อกราฟ , c เปลี่ยนสีกราฟ
plt.plot(x,y2,label = 'cos(x)',c='r')
plt.legend(loc = 1) ## โชว์ label พร้อมกับกับตั้งตำเเหน่ง loc (1-10)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Compare between Sin and Cos Graph')
plt.xlim(0,2*3.14) ## limit เเกน x
plt.ylim(-1,1)## limit เเกน y
plt.grid() ## เพิ่มเส้น grid
plt.show()
```

```python
########## Layout (กราฟหลายกราฟในภาพเดียว)
x = np.arange(0,20,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
fig = plt.figure()
ax1 = fig.add_subplot(211,title ='Pic.1')
ax1.set_xlabel('x')
ax1.set_xlabel('y')
ax1.plot(x,y1,'m')
### 2 อัตราส่วนเเกน x , 1 อัตราส่วนเเกน y ,1 รูปที่
ax2 = fig.add_subplot(212,title ='Pic.2')
ax2.set_xlabel('x')
ax2.set_xlabel('y')
ax2.plot(x,y2,'r')
plt.tight_layout() ## เลื่อนตัวหนังสือที่ทับกัน
plt.show()
```

```python
x = np.arange(0,20,0.1)
y1 = np.sin(x)
y=x**2
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1]) ## เปอร์เซ็นความกว้างของกราฟย่อย นับตามลำดับ จากซ้าย-ขวา, ขึ้น-ลง, ยาว,กว้าง จากจุด (0,0)
ax1.plot(x,y,'r')

ax1 = fig.add_axes([0.1,0.55,0.4,0.4]) ## เปอร์เซ็นความกว้างของกราฟย่อย นับตามลำดับ จากซ้าย-ขวา, ขึ้น-ลง, ยาว,กว้าง จากจุด (0,0)
ax1.plot(x,y1,'m')
plt.show()
fig.savefig('Multiple graph Mini-scale',dpi = 300) ## save ไฟล์ กราฟ ลงเครื่อง
```

```python
####################### Scatter plot ดูการกระจายตัว
x = np.arange(0,10)
y = np.random.randint(1,150,10)
y2 = np.random.randint(10,110,10)
## สร้าง graph ขนาด 5x5
plt.figure(figsize=[5,5])
plt.scatter(x,y,c='r')
plt.scatter(x,y2,c='g')
plt.title('Scatter Plot')
plt.show()
```

```python
####################### Bar plot ดูการกระจายตัว ******จำนวน x,y ต้องเท่ากัน
x = np.arange(1,13)
y = np.random.randint(1,150,12)
y = np.sort(y) ## เรียงค่าเเกน y
plt.figure(figsize=[5,5])
plt.bar(x,y)
plt.title('Bar plot Model in 1 year')
plt.show()
```

```python
##### Histrogram ดูการกระจายตัวของข้อมูล
x = np.random.randint(1,100,100)
plt.hist(x,bins=10) ## bins = ช่วงของข้อมูล
plt.show()
```

```python
#### Pie plot สัดส่วนของข้อมูล
x = np.random.randint(1,100,5)
label1 = ['Benz','BMW','Toyota','Mazda','Nissan']
plt.figure(figsize=[5,5])
plt.pie(x,labels =label1,explode = [0,0.1,0,0,0.1]) ## explode ยื่นออกมาตามลำดับข้อมูล
plt.title('Sale car in thailand')
plt.show()
```
